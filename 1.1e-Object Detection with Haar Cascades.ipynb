## Face Detection with Haar Cascades

```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nadia = cv2.imread('images/Nadia_Murad.jpg',0)
denis = cv2.imread('images/Denis_Mukwege.jpg',0)
solvay = cv2.imread('images/solvay_conference.jpg',0)
```

```python
plt.imshow(nadia,cmap='gray')
plt.imshow(denis,cmap='gray')
plt.imshow(solvay,cmap='gray')
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    
  
    face_img = img.copy()
    # scaleFactor refers to the image pyramid
    # minNeighbors refers to the number of feature neighbors to confirm a 
    face_rects = face_cascade.detectMultiScale(face_img,scaleFactor=1.2, minNeighbors=5) 
    
    for (x,y,w,h) in face_rects: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 10) 
        
    return face_img
```

```python
result = detect_face(solvay)
plt.imshow(result,cmap='gray')
```
